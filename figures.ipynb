```python
from Gaussians import FTL_online, FTL, MWUA_FTL
import pickle


names = ['10k','_ill', '_well_3k', '_well_4kBut3kalg']  #'5k']
methods = ['ftl','mwua']



class regret:
    def __init__(self, name):
        self.ftl = pickle.load(open('ftl' + name + ".p","rb"))
        self.mwua = pickle.load(open('mwua' + name + ".p","rb"))
    
    def plot(self):
        pass



```

```python
import matplotlib.pyplot as plt
import math
import numpy as np

# --- mnist ---

mnist = regret(names[0])

x = np.array(range(1,len(mnist.mwua[0])+1))
mwua = np.array(mnist.mwua[0])/784
ftl = np.array(mnist.ftl[0])/784


fig, ax = plt.subplots()
ax.plot(np.log2(x), np.log2(mwua), label='mwua')
ax.plot(np.log2(x), np.log2(ftl), label='ftl')
ax.plot(np.log2(x), np.log2(x)*0.20, label='0.2x')
ax.plot(np.log2(x), np.log2(x)*0.33-0.5, label='0.33x-0.5')
ax.set_xlabel('log2(t)')
ax.set_ylabel('log2(regret)')
ax.set_title('mnist10 log2(regret) vs. log2(t)')
ax.legend()
fig.show()

fig2, ax2 = plt.subplots()
ax2.plot(np.log2(x), mwua, label='mwua-logx')
ax2.plot(np.log2(x), ftl, label='ftl-logx')
ax2.plot(np.log2(x), np.log2(x)*1.5-6, label='1.5x-6')
ax2.plot(np.log2(x), np.log2(x)*0.5-0.5, label='0.5x-0.5')
ax2.set_xlabel('log2(t)')
ax2.set_ylabel('regret')
ax2.set_title('mnist10 regret vs. log(t)')
ax2.legend()
fig2.show()

fig3, ax3 = plt.subplots()
ax3.plot(mwua, label='mwua')
ax3.plot(ftl, label='ftl')
ax3.plot(x, np.log2(x), label='log2')
ax3.set_xlabel('t')
ax3.set_ylabel('regret')
ax3.set_title('mnist10 regret vs. t')
ax3.legend()
fig3.show()


```
